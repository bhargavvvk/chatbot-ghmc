```python
import os
import time
import speech_recognition as sr
from gtts import gTTS
from transformers import GPT2Tokenizer, TFGPT2LMHeadModel
import numpy as np
import tensorflow as tf

# Beginning of the AI
class ChatBot():
    def __init__(self, name):
        print("----- starting up", name, "-----")
        self.name = name 
        self.text = ""

    def speech_to_text(self):
        recognizer = sr.Recognizer()
        with sr.Microphone() as mic:
            print("listening...")
            audio = recognizer.listen(mic)
            self.text = "ERROR"
        try:
            self.text = recognizer.recognize_google(audio)
            print("me --> ", self.text)
        except:
            print("Me  -->  ERROR")

    def wake_up(self, text):
        return True if self.name in text.lower() else False

    @staticmethod
    def text_to_speech(text):
        print("bruna-->", text)
        speaker = gTTS(text=text, lang="en", slow=False)
        speaker.save("res.mp3")
        os.system('afplay res.mp3')  # Use 'afplay' on Mac or 'start' on Windows
        time.sleep(2)  # Give time for the audio to play
        os.remove("res.mp3")

# Execute the AI
ai = ChatBot(name="bruna")

# Load the DialoGPT model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("microsoft/DialoGPT-medium", padding_side="left")
model = TFGPT2LMHeadModel.from_pretrained("microsoft/DialoGPT-medium")
os.environ["TOKENIZERS_PARALLELISM"] = "true"

ex = True
while ex:
    ai.speech_to_text()
    if ai.wake_up(ai.text) is True:
        res = "Hello, I am Bruna the Chatbot. What can I do for you?"
        ai.text_to_speech(res)
        ai.text = ""
    elif any(i in ai.text for i in ["exit", "close"]):
        res = np.random.choice(["Tata", "Have a good day", "Bye", "Goodbye", "Hope to meet soon", "Peace out!"])
        ai.text_to_speech(res)
        ex = False
    else:
        if ai.text == "ERROR":
            res = "Sorry, come again?"
        else:
            # Use the most recent user input without context
            input_text = ai.text
            input_ids = tokenizer.encode(input_text + tokenizer.eos_token, return_tensors="tf")

            # Generate a response with adjusted parameters
            chat_output = model.generate(
                input_ids,
                max_length=100,
                pad_token_id=tokenizer.eos_token_id,
                temperature=0.7,
                top_k=50,
                top_p=0.9
            )
            res = tokenizer.decode(chat_output[0], skip_special_tokens=True).strip()

        ai.text_to_speech(res)

print("----- Closing Bruna ----")
```
